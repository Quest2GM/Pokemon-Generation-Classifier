```
# Requests for fetching html of website
import requests
import urllib
from bs4 import BeautifulSoup

# Make the request to a url and extract soup content
r = requests.get('https://pokemon-trainer.com/wiki/sprite-sun-moon-shiny/').content
soup = BeautifulSoup(r)
```

```
# All URLs:
    # Gen 1:
        # Red/Blue:                     https://pokemon-trainer.com/wiki/sprite-red-blue/
        # Yellow:                       https://pokemon-trainer.com/wiki/sprite-yellow/
    # Gen 2:
        # Silver:                       https://pokemon-trainer.com/wiki/sprite-silver/
        # Gold:                         https://pokemon-trainer.com/wiki/sprite-gold/
    # Gen 3:
        # Ruby/Sapphire:                https://pokemon-trainer.com/wiki/sprite-ruby-sapphire/
        # Fire-Red/Leaf-Green:          https://pokemon-trainer.com/wiki/sprite-fire-red-leaf-green/
    # Gen 4/5:
        # Platinum:                     https://pokemon-trainer.com/wiki/sprite-platinum/
        # Platinum Shiny:               https://pokemon-trainer.com/wiki/sprite-platinum-shiny/
        # Heart-Gold/Soul-Silver:       https://pokemon-trainer.com/wiki/sprite-heart-gold-soul-silver/
        # Heart-Gold/Soul-Silver Shiny: https://pokemon-trainer.com/wiki/sprite-heart-gold-soul-silver-shiny/
        # Black/White:                  https://pokemon-trainer.com/wiki/sprite-black-white/
        # Black/White Shiny:            https://pokemon-trainer.com/wiki/sprite-black-white-shiny/
    # Gen 6/7:
        # XY Oras:                      https://pokemon-trainer.com/wiki/sprite-xy-oras/
        # XY Oras Shiny:                https://pokemon-trainer.com/wiki/sprite-xy-oras-shiny/
        # Sun/Moon:                     https://pokemon-trainer.com/wiki/sprite-sun-moon/
        # Sun/Moon Shiny:               https://pokemon-trainer.com/wiki/sprite-sun-moon-shiny/
```

```
def download_image(url):
    name = ''.join(filter(str.isdigit, url))
    fullname = name +".png"                   # Name sprites by their Pokedex number
    urllib.request.urlretrieve(url, fullname) 
```

```
# Find the element on the webpage
images = soup.findAll('img')

for i in images:
    if ("sprite") in (str)(i.get("src")):
        download_image((str)(i.get("src")))
```
